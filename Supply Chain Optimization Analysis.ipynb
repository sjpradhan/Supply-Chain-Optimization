This dataframe exclusively holds customer demographic information, focusing on key identifiers and characteristics. The order_info dataframe is dedicated to information pertaining to orders. The order_item_info dataframe is dedicated to detailed information about individual order items. It covers attributes such as Order Item Id, Order Item Cardprod Id, Order Item Discount, Order Item Discount Rate, Order Item Product Price, Order Item Profit Ratio, Order Item Quantity, Sales, Order Item Total, and Order Profit Per Order. This separation provides a granular view of the components contributing to each order.

__Location Information Dataframe (location_info):__

Focused on geographical aspects, the location_info dataframe includes attributes like Product Card Id, Latitude, Longitude, Market, Department Id, and Department Name. This separation facilitates the examination of spatial relationships and geographical distribution of products.

__Sales Information Dataframe (sales_info):__

The sales_info dataframe concentrates on attributes related to sales transactions, such as Type, Benefit per order, Sales per customer, Delivery Status, Late_delivery_risk, Category Id, and Category Name. This separation enables a focused analysis of sales-related metrics and transactional details.

## Hypothesis Test:
H0 = Mean Of Discount On Loss Making Orders - Mean Of Discount On Pofit Making Orders <=0
<br>

H1 = Mean Of Discount On Loss Making Orders - Mean Of Discount On Pofit Making Orders > 0
<br>

Alpha = 0.05

__Two common tests for comparing means are the t-test and the Mann-Whitney U test (a non-parametric test)__
<br>

__T-Test:__
The t-test is suitable when the data is approximately normally distributed, and comparing means between two groups.
<br>

__Assumptions:__
Normally distributed data.
Homogeneity of variances (variances are roughly equal between the groups).
<br>

__Mann-Whitney U Test:__
The Mann-Whitney U test is a non-parametric test that doesn't assume a specific distribution. It's suitable when the data is not normally distributed, or when data is ordinal or interval data.

__Assumptions:__ 
Fewer assumptions compared to the t-test. It does not assume normality or equal variances.
<br>

__Choosing Between Them:__

If the data meets the assumptions of the t-test, it is generally more powerful (better ability to detect true effects).

If the data violates the assumptions of the t-test, or if unsure about the distribution of the data, the Mann-Whitney U test might be a more robust option.

__Steps:__

Conduct normality tests or visually inspect the data to check for normality.
If normality is met and variances are roughly equal, consider a t-test.
If normality is not met or you have ordinal data, consider the Mann-Whitney U test.

## Checking Distribution Of Data:

__For the "Loss" group:__

Shapiro-Wilk test statistic (stat_loss) is approximately 0.945
p-value (p_value_loss) is 0.0

__For the "Profit" group:__

Shapiro-Wilk test statistic (stat_profit) is approximately 0.945
p-value (p_value_profit) is 0.0

In both cases, the p-values are very close to zero, suggesting strong evidence against the null hypothesis that the data is normally distributed. Therefore, based on the Shapiro-Wilk test, we may conclude that the "Order Item Discount Rate" for both "Loss" and "Profit" groups does not follow a normal distribution.

```python
from scipy import stats
stat_loss, p_value_loss = stats.shapiro(new_df[new_df["Net Income"] == 'Loss']['Order Item Discount Rate'])
stat_profit, p_value_profit = stats.shapiro(new_df[new_df["Net Income"] == 'Profit']['Order Item Discount Rate'])

# Check the normality of each group using the Shapiro-Wilk test
plt.figure(figsize=(6, 3))
sns.histplot(new_df[new_df["Net Income"] == 'Loss']['Order Item Discount Rate'], label='Loss', kde=True, color='red', alpha=0.5)
sns.histplot(new_df[new_df["Net Income"] == 'Profit']['Order Item Discount Rate'], label='Profit', kde=True, color='blue', alpha=0.5)

plt.legend()
plt.show()
```

```python
print(f'stat_loss: {stat_loss}')
print(f'stat_profit: {stat_profit}')
print(f'p_value_loss: {p_value_loss}')
print(f'p_value_profit: {p_value_profit}')
```

```python
# Perform the t-test if both groups are approximately normally distributed
if p_value_loss > 0.05 and p_value_profit > 0.05:
    t_stat, p_value = stats.ttest_ind(new_df[new_df["Net Income"] == 'Loss'][column_name],
                                       new_df[new_df["Net Income"] == 'Profit'][column_name],
                                       alternative='greater')
    print(f'The t-test p-value is: {p_value}')
else:
    print(f'One or both groups do not appear to be normally distributed. Consider non-parametric tests or transformations.')
```

If the assumption of normality is violated, we might consider using non-parametric tests or other robust methods. In our case, since the Shapiro-Wilk test indicates a departure from normality, I should use the Mann-Whitney U test, which is a non-parametric alternative to the independent samples t-test. The Mann-Whitney U test does not assume normality and is suitable for comparing distributions of two independent groups.

Null Hypothesis (H0): The mean discount on "Loss" making orders is less than or equal to the mean discount on "Profit" making orders.

Alternative Hypothesis (H1): The mean discount on "Loss" making orders is greater than the mean discount on "Profit" making orders.

Since the p-value is greater than 0.05, there is not enough evidence to reject the null hypothesis. It suggests that there is insufficient evidence to conclude that the mean discount on "Loss" making orders is significantly greater than the mean discount on "Profit" making orders.

In simpler terms, based on the Mann-Whitney U test, we do not have enough evidence to claim that there is a significant difference in the mean discount between the "Loss" and "Profit" groups.

```python
from scipy.stats import mannwhitneyu

loss_data = new_df[new_df["Net Income"] == 'Loss']['Order Item Discount Rate']
profit_data = new_df[new_df["Net Income"] == 'Profit']['Order Item Discount Rate']

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(loss_data, profit_data, alternative='less')

# Print the results
print(f"Mann-Whitney U statistic: {statistic}")
print(f"P-value: {p_value}")
```
